```python
import numpy as np
import matplotlib.pyplot as plt

#Konstanten a,b
a = 1
b = 2  

#Potential V(x,y)
def V(x,y):
    return a*x**2+b*y**2

#Gitter
x = np.linspace(-20, 20,10)
y = np.linspace(-20, 20,10)
x_vec, y_vec = np.meshgrid(x, y)
z = V(x_vec, y_vec)

#Kraftfeld F = -grad(V)
F_x, F_y=np.gradient(-z)

#Plot
plt.contour(x_vec, y_vec, z, 5)
plt.quiver(x,y, F_y,F_x)
plt.show()
```
